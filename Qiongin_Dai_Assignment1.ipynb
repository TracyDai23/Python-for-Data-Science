# 10)	Top 5 highest OBP (on base percentage) with at least 500 PAs in 1977?  (Dataframe)

```python
dfc=df.fillna(0)
dfc=dfc.groupby(['playerID','nameFirst','nameLast','yearID','birthYear','age'],as_index=False).sum()
```

```python
dfc['PA']=dfc['AB']+dfc['BB']+dfc['IBB']+dfc['SH']+dfc['SF']
```

```python
dfc['OBP']=(dfc['H']+dfc['BB']+dfc['IBB']+dfc['SH']+dfc['SF'])/dfc['AB']
```

```python
df10 = dfc[(dfc['yearID']==1977)&(dfc['PA']>=500)]
df10.nlargest(5,'OBP')
```

# BONUS:Using a line graph please graph the average HRs per AB (think about this) per season
